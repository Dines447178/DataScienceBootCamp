```python
print ('******BONUS ELIGIBILITY CHECK******')
salary =float (input('PLEASE ENTER YOUR SALARY:'))
years =float (input('PLEASE ENTER THE YEAR OF SERVICE:'))
if years > 5 :
  print ('CONGRATULATION YOU ARE ELIGIBLE FOR 5% BONUS')
  bonus= 1.05 *salary
  print ('YOUR BONUS AMOUNT IS', bonus)
else:
  print ('NOT ELIGIBLE!! SORRY YOU NEED TO SERVE AT LEAST 5 YEARS TO RECEIVE BONUS')
```
